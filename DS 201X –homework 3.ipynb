```python
import pandas as pd
import seaborn as sns
```

```python
#read in the csv to a dataframe
df = pd.read_csv("USvideos.csv/USvideos.csv")
```

```python
#check the top couple entries in the dataframe
df.head()
```

```python
#look at how the data is described
df.describe()
```

```python
#we drop any duplicate values in the dataframe
df.drop_duplicates()
```

```python
#check to see if any duplicates were removed
df.describe()
```

```python
#check the correlation between each of the categories
corr = df.corr()
corr.style.background_gradient()
```

```python
#check the correlation between the views and likes by category
df.plot.scatter(x='likes',y='views',c='category_id',colormap='viridis')
```

```python
#grab the top 100 entries based off of the views
df_top100views = df.nlargest(100, 'views')
```

```python
#grabbing the top ten entries based off of likes and views
df.nlargest(10, ['likes','views'])
```

```python
#create new dataframe from the columns within the main dataframe
data1 = df[['views','likes','category_id']]
data1
```

```python
#use the new dataframe to create a pairplot to find correlation 
sns.pairplot(data1, hue='category_id')
```

```python
#create new dataframe based off of the top 100 entries of views
data2 = df_top100views[['views','likes','category_id']]
```

```python
#plot the pairplot
sns.pairplot(data2, hue='category_id')
```
